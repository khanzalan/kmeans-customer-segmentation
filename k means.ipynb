```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

np.random.seed(42)
customers = np.vstack([
    np.random.normal([200,  5],  [30, 1], (30, 2)),   # Low spenders
    np.random.normal([500,  15], [40, 2], (30, 2)),   # Mid spenders
    np.random.normal([900,  30], [50, 3], (30, 2)),   # High spenders
])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(customers)
centers = kmeans.cluster_centers_

# Q13: Scatter Plot
colors = ['#E74C3C', '#2ECC71', '#3498DB']
names  = ['Low Spenders', 'Mid Spenders', 'High Spenders']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Raw data
axes[0].scatter(customers[:, 0], customers[:, 1], color='gray', alpha=0.6, s=60)
axes[0].set_title('Before Clustering', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Annual Spend ($)')
axes[0].set_ylabel('Purchase Frequency')
axes[0].grid(True, alpha=0.3)

# Right: Clustered
for i in range(3):
    mask = labels == i
    axes[1].scatter(customers[mask, 0], customers[mask, 1],
                    color=colors[i], label=names[i], alpha=0.75, s=70)
axes[1].scatter(centers[:, 0], centers[:, 1],
                color='black', marker='X', s=220, zorder=5, label='Centroids')
axes[1].set_title('After K-Means Clustering (k=3)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Annual Spend ($)')
axes[1].set_ylabel('Purchase Frequency')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.suptitle('Customer Segmentation using K-Means', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('kmeans_clusters.png', dpi=120, bbox_inches='tight')
plt.show()
print("✅ Scatter plot saved as 'kmeans_clusters.png'")

for i, c in enumerate(centers):
    print(f"  Cluster {i+1} ({names[i]}): Avg Spend=${c[0]:.0f}, Avg Frequency={c[1]:.1f}")
```
